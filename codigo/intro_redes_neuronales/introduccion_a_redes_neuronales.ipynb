**Suma de Tensores**  
Existen dos formas de sumar tensores. Una usando el simbolo de suma, y otra usando la funcion add de pytorch.

````
x = 2*torch.ones(2,2)
y = torch.ones(2,2)

print(x)
print(y)

print('\n Suma : \n', x + y)
print('\n Suma : \n', torch.add(x,y))
````

**Suma inplace**  
Tambien puedes modificar un tensor con suma usando add_.

+ En general todas las operaciones inplace de pytorch se realizan con el _ (guion bajo).

**Cuidado con los apuntadores**:
Observa como al modificar z en el siguiente snippet, y se ve modificada.

````
print(y)
z = y
z.add_(x)
print()
print(y)
````

**Copiar tensores**
Existen varias formas de copair tensores. La mas rapida es usar .clone().detach().  
Si deseas saber mas sobre el tema revisa el post de stackoverflow:  
https://stackoverflow.com/questions/55266154/pytorch-preferred-way-to-copy-a-tensor 

````
print(y)
z = y.clone().detach()
z.add_(x)
print()
print(y)
````

### Reshape de Tensores

El metodo view nos permite modificar la forma de nuestros tensores.    

+ Si utilizamos el argumento -1 aplanara el vector a lo largo de esa dimension.  

````
x = torch.rand(4,4)
print(x)

print('\n Tensor Unidimensional : \n')
print(x.view(-1))
````

Si acompannanos al argumento menos 1 con el tamano que queremos en la otra dimension, pytorch lo calculara automaticamente.  

+ 8 columnas -> **(-1,8)** 

+ 8 renglones -> **(8, -1)** 

````
print(x)
print('\n Tensor de 8 columnas : \n')
print(x.view(-1, 8))

print('\n Tensor de 8 renglones : \n')
print(x.view(8, -1))
````

# Red neuronal 101

## Red Neuronal con objetos


Construyamos nuestra primera Red Neuronal usando objetos.

**Documentacion**
nn.Lienear: https://pytorch.org/docs/stable/generated/torch.nn.Linear.html 


image.png

**Modificacion**: El metodo que observas esta modificado respecto a lo que observas en las notas de clase. Ahora es mas general y acepta diferentes neuronas entre las capas.

```
in_n= numero de datos o p 
hidden_n = numero de neuronas en la capa oculta
out_n= numero de neuronas en la capa de salida
```



````
from torch import nn

class RedNeuronal(nn.Module):

    def __init__(self, in_n=784, hidden_n=256, out_n=1):
        super(RedNeuronal, self).__init__()
        
        # W0:= Pesos entrada -> oculta 
        # nn.Linear(neuronas de entrada, neuronas ocultas, bias=True)
        self.hidden = nn.Linear(in_n, hidden_n)

        # W1:= Pesos oculta -> salida 
        # nn.Linear(neuronas ocultas, neuronas salida, bias=True)
        self.output = nn.Linear(hidden_n, out_n)
        
        # Funciones de Activacion 
        # Sigma 1 [capa oculta]
        self.sigmoid = nn.Sigmoid()
        # Sigma 2 [capa de salida]
        self.softmax = nn.Softmax(dim=1)
        
    def forward(self, x):
        # Flujo o feedforward de nuestra red
        
        # Multiplicacion de X0 por W0
        x = self.hidden(x)
        # Evaluar Sigma uno / Obtener X1
        x = self.sigmoid(x)

        # Multiplicacion de X1 por W1
        x = self.output(x)

        # Evaluar en Sigma 2 / Obtener X2
        x = self.softmax(x)
        return x
````

### Hagamos un feedforward    
**Nota**: Al construir nuestro objeto RedNeuronal los pesos y bias son inicializados automaticamente. Esta eleccion es educada, y lo veremos mas adelante.

image.png

````
# Establecer semilla
torch.manual_seed(345)

# 28x28 = 784 Realmente es el tamano de una imagen
x0 = torch.rand((1, 784))
model_nn = RedNeuronal(in_n=784, hidden_n=256, out_n=1)
resultado = model_nn(x0)
print(resultado)
````

**Como sabemos si es correcto lo que hicismos?**  
Que te parece si evaluamos nuestro codigo revisando el feedforward.  
+ Hagamos una red mas pequenna.  
+ **Cambiemos la funcion softmax por la identidad para evaluar facilmente.**


````
# Establecer semilla
torch.manual_seed(345)

# Hyperparamteros de nuestra red
# p, n0 o numero de variables
n_in = 3
# n1, o numero de neuronas ocultas
n_hid = 5
# n2, o numero de neuronas de salida
n_out = 1

x0 = torch.rand((1, n_in))

model_nn = RedNeuronal(in_n=n_in, hidden_n=n_hid, out_n=n_out)
# Eliminemos la funcion softmax de la ultima capa
# Ahora sigma 2 sera la identidad, para que nos sea mas facil evaluar
model_nn.softmax = nn.Identity()
resultado = model_nn(x0)
print(resultado)
````


### Evaluando feedforward

Veamos los pesos W0 que conectan a nuestra capa de entrada con los de salida.
+ Podemos hacer uso del metodo ```
.weight ``` Que tienen nuestros pesos definidos por ``` nn.Linear ```.  

+ **Documentacion de Linear** :
https://pytorch.org/docs/stable/generated/torch.nn.Linear.html





````
# Veamos si nuestra red cargo los pesos correctamente.
print(model_nn.hidden.weight)
````

**Guardemos nuestros pesos y evaluemos que nuestro codigo es correcto**.

````
# Entrada a la capa oculta
W0 = model_nn.hidden.weight
b0 = model_nn.hidden.bias[None,:]
# No olvides utilizar None para hacer las dimensiones compatibles
print('Dim W0:   ', W0.shape)
print('Dim b0:   ', b0.shape)

print()

# Entrada a la capa de salida
W1 = model_nn.output.weight
# No olvides utilizar None para hacer las dimensiones compatibles
b1 = model_nn.output.bias[None,:] 
print('Dim W1:   ', W1.shape)
print('Dim b1:   ', b1.shape)

# Revisemos la dimension de x0
print('\nDim de x0 : ',x0.shape)
````

**Las dimensiones son correctas**?  
**Si, pero las matrices estan transpuestas**.

+ **Evaluemos la entrada a la capa oculta**   
+ **Apliquemos sigma 1 que es sigmoidal**

````
# Entrada la capa oculta
input1 = torch.matmul(x0, W0.transpose(0,1))
input1 = input1 + b0
print(input1.shape)

# salida de la capa oculta
sigma_1 = nn.Sigmoid()
hidden_out = sigma_1(input1)
hidden_out.shape
````

+ **Evaluemos la entrada a la capa de salida**   
+ **Apliquemos sigma 2 que es softmax**

````
# Entrada a la capa de salida
input2 = torch.matmul(hidden_out, W1.transpose(0,1))
input2 = input2 + b1
print(input2.shape)

# Output de nuestra red
sigma_2 = nn.Identity()
y_out = sigma_2(input2)
y_out[0]
````

**Parece que nuestro objeto funciona**

**Eficiencia**: Probablemente seria mas eficiente transponer las matrices desde antes. En este caso lo hicimos asi para recordar esa diferencia de notacion.

## Red Neuronal con Sequential

Recostruyamos nuestra red neuronal utilizando sequential.

````
# Hyperparamteros de nuestra red
# p, n0 o numero de variables
datos = 784
# n1, o numero de neuronas ocultas
neuronas_ocultas = [256]
# n2, o numero de neuronas de salida
neuronas_salida = 1

# Red Neuronal con arquitectura feedforward
model_s = nn.Sequential(
    # X0 x W0 
    nn.Linear(datos, neuronas_ocultas[0]), 
    # X1 = sigma_1(W0 x X0)
    nn.Sigmoid(),
    # X1 x W1
    nn.Linear(neuronas_ocultas[0], neuronas_salida),
    # Y = X2 = sigma_2(W1 x X1)
    nn.Softmax(dim=1))

print(model_s)
````

````
# Establecer semilla
torch.manual_seed(345)

# 28x28 = 784 Realmente es el tamano de una imagen
x0 = torch.rand((1, 784))
model_s(x0)
````
